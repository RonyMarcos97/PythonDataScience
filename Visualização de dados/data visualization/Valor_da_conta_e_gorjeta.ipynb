```python
import seaborn as sns
import pandas as pd
```

```python
dados = pd.read_csv('gorjetas.csv', sep=';')
```

```python
dados.columns
```

```python
valor_gorjeta = sns.scatterplot(x='valor_total', y='gorjeta', data=dados)
```

### Visualmente, o valor da gorjeta aumenta conforme aumenta o valor da conta

```python
print('A base de dados contém {} registros \n'.format(dados.shape[0]))
print('Registros não nulos')
dados.count()
```

## Criando o campo porcentagem

```python
dados.head(1)
```

```python
dados['porcentagem'] = dados['gorjeta'] / dados['valor_total']
```

```python
dados.head()
```

```python
dados.porcentagem = dados.porcentagem.round(2)
```

```python
dados.head()
```

```python
porcentagem_conta = sns.scatterplot(x='valor_total', y='porcentagem', data=dados)
```

### Visualmente, o valor da conta não é proporcional ao valor da gorjeta

```python
porcentagem_conta_linha = sns.relplot(x='valor_total', y='porcentagem', kind='line', data=dados)
```

```python
sns.lmplot(x='valor_total', y='porcentagem', data=dados)
```

### Podemos analisar que o valor da gorjeta vai caindo

```python
dados.to_csv('gorjetas.csv', sep=';', index=False)
```
