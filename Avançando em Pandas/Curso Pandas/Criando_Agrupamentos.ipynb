## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
dados['Valor'].mean()
```

```python
bairros = list(dados['Bairro'].values)
```

```python
bairros
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data.Valor.mean()))
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': ' Mínimo', 'max': 'Máximo', 'sum': 'Total'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
plt.show()
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Maxímo do Aluguel por Bairro', {'fontsize': 22})
plt.show()
```

```python
fig = grupo_bairro['Valor'].min().plot.bar(color='blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Minimo do Aluguel por Bairro', {'fontsize': 22})
plt.show()
```
